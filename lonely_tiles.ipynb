```python
a = ('a', {1, 5, 7, 8})
b = ('b', {1, 5, 6, 7, 8})
c = ('c', {4})
d = ('d', {5, 8})
e = ('e', {5, 6, 8, 9})
f = ('f', {2})
g = ('g', {3})
h = ('h', {1, 5, 8})
i = ('i', {5, 8})
```

```python
tiles = [a, b, c, d, e, f, g, h, i]
```

```python
tiles = sorted(tiles, key=lambda a: len(a[1]))
```

```python
def remove_lonely_tiles(tiles: tuple[str, set]) -> tuple[str, set]:
    output = []
    union = set()
    remove = set()
    for i, tile in enumerate(sorted(tiles, key=lambda a:len(a[1])), 1):
        new_union = union | tile[1]
        if (len(new_union - union) == 1):
            output.append((tile[0], new_union - union))
        else:
            output.append((tile[0], tile[1]-remove))
        if len(new_union) == i:
            remove = new_union    
        union = new_union

    return sorted(output, key=lambda a: len(a[1]))
```

```python
def remove_all_lonely_tiles(tiles: tuple[str, set]) -> tuple[str, set]:
    prev_output = set()
    output = tiles
    while (output != prev_output):
        prev_output = output
        output = remove_lonely_tiles(output)
    print(output)
    
```

```python
remove_all_lonely_tiles(tiles)
```
